```python
import pandas as pd
from sklearn.model_selection import cross_validate, KFold
from sklearn.linear_model import ElasticNet
from sklearn.metrics import make_scorer, mean_squared_error
import numpy as np

# Carregar dados
df = pd.read_csv("regressao_Q1.csv")

# Separar variáveis preditoras e alvo
# Ajuste o nome da variável target conforme o arquivo
X = df.drop(columns=["target"])
y = df["target"]

# Modelo Elastic Net
model = ElasticNet(
    alpha=1.0,
    l1_ratio=0.01,
    random_state=42
)

# Validação cruzada com 5 folds
cv = KFold(
    n_splits=5,
    shuffle=True,
    random_state=42
)

# RMSE como métrica
scoring = {
    "mse": make_scorer(mean_squared_error, greater_is_better=False)
}

results = cross_validate(
    model,
    X,
    y,
    cv=cv,
    scoring=scoring,
    return_train_score=True
)

# Como o sklearn retorna negativo para erro quando greater_is_better=False
train_mse = -results["train_mse"]
val_mse = -results["test_mse"]

# Média dos erros quadráticos
mean_train_mse = train_mse.mean()
mean_val_mse = val_mse.mean()

print("MSE em treino por fold:", train_mse)
print("MSE em validação por fold:", val_mse)

print("Média MSE treino:", mean_train_mse)
print("Média MSE validação:", mean_val_mse)
```
