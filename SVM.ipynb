```python
import pandas as pd
import numpy as np

from sklearn.svm import SVR
from sklearn.model_selection import cross_validate, KFold
from sklearn.metrics import make_scorer, mean_squared_error
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

# Carregar dados
df = pd.read_csv("regressao_Q2.csv")

# Ajuste o nome da variável alvo conforme o arquivo
X = df.drop(columns=["target"])
y = df["target"]

# Modelo SVM com kernel linear e C = 0.001
model = Pipeline([
    ("scaler", StandardScaler()),
    ("svm", SVR(kernel="linear", C=0.001))
])

# Validação cruzada sequencial: shuffle=False
cv = KFold(
    n_splits=5,
    shuffle=False
)

# MSE
scoring = {
    "mse": make_scorer(mean_squared_error, greater_is_better=False)
}

results = cross_validate(
    model,
    X,
    y,
    cv=cv,
    scoring=scoring,
    return_train_score=True
)

# Converter para positivo
train_mse = -results["train_mse"]
val_mse = -results["test_mse"]

print("MSE treino por fold:", train_mse)
print("MSE validação por fold:", val_mse)

print("Média MSE treino:", train_mse.mean())
print("Média MSE validação:", val_mse.mean())
```
