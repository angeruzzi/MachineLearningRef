```python
import pandas as pd

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import KFold, cross_validate
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

df = pd.read_csv("classificacao_Q2.csv")

# Ajuste o nome da variável alvo conforme o arquivo
X = df.drop(columns=["target"])
y = df["target"]

model = Pipeline([
    ("scaler", StandardScaler()),
    ("logreg", LogisticRegression(
        penalty="l2",
        C=0.1,
        solver="liblinear",
        max_iter=1000
    ))
])

cv = KFold(
    n_splits=10,
    shuffle=False
)

scores = cross_validate(
    model,
    X,
    y,
    cv=cv,
    scoring="roc_auc",
    return_train_score=True
)

print("AUC treino por fold:", scores["train_score"])
print("AUC validação por fold:", scores["test_score"])

print("Média AUC treino:", scores["train_score"].mean())
print("Média AUC validação:", scores["test_score"].mean())
```
